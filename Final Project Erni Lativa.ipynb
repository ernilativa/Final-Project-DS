# Data Understanding & Penentuan Goals

Dataset menyediakan informasi yang sangat berharga mengenai pola pembelian konsumen di industri ritel. Dataset ini mencakup berbagai kolom yang menggambarkan transaksi konsumen, produk yang dibeli, serta karakteristik pelanggan. Data yang ada sangat penting untuk memahami tren konsumen, perilaku pembelian, dan untuk merancang strategi pemasaran yang lebih terfokus.


Fitur Utama yang terdapat dalam Dataset:

**Detail Transaksi:**

1. Tanggal Pembelian: Waktu kapan transaksi dilakukan.
2. Nilai Transaksi: Jumlah uang yang dihabiskan untuk setiap pembelian.
3. Kategori Produk: Jenis produk yang dibeli (misalnya, pakaian, elektronik, dll.).
4. Metode Pembayaran: Cara pembayaran (misalnya, kartu kredit, uang tunai, atau pembayaran digital).


**Informasi Pelanggan:**

1. Kelompok Usia: Kategori umur pelanggan (misalnya, 18-25 tahun, 26-35 tahun, dll.).
2. Jenis Kelamin: Identifikasi jenis kelamin pelanggan (laki-laki atau perempuan).
3. Lokasi: Tempat tinggal pelanggan atau area geografis.
4. Status Loyalitas: Apakah pelanggan terdaftar sebagai anggota program loyalitas.


**Perilaku Belanja:**

1. Frekuensi Pembelian: Seberapa sering pelanggan berbelanja.
2. Pengeluaran Rata-rata per Transaksi: Rata-rata jumlah uang yang dikeluarkan setiap transaksi.
3. Tren Musiman: Perubahan kebiasaan belanja berdasarkan musim atau waktu tertentu (misalnya, saat promosi atau musim liburan).


**Masalah**
Menganalisis perilaku pembelian pelanggan untuk mengidentifikasi apakah pelanggan berisiko berhenti membeli produk.

**Tujuan Penggunaan Dataset:**
Dataset ini sangat berguna bagi data scientist, analis, dan pemasar untuk memprediksi permintaan produk di masa depan serta mengidentifikasi pelanggan yang berisiko mengalami churn (berhenti berbelanja). Dengan demikian, dataset ini dapat membantu dalam merencanakan strategi pemasaran yang lebih efektif dan meningkatkan retensi pelanggan


**Manfaat Dataset:**
Baik untuk analisis eksplorasi data (EDA), membuat visualisasi, atau melatih model machine learning, dataset ini menyediakan wawasan berharga yang mendukung pengambilan keputusan berbasis data dalam sektor ritel.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('shopping_trends.csv')
df
```

**Detail informasi dalam setiap kolom dataset:**

1. Kolom Customer ID: Berisi ID unik pelanggan, dengan nilai unik dari 1 hingga 3900. Biasanya hanya digunakan sebagai pengenal.

2. Kolom Age: Menunjukkan usia pelanggan, dengan nilai unik berkisar antara 18 hingga 70 tahun.

3. Kolom Gender: Berisi jenis kelamin pelanggan, dengan nilai unik berupa 'Male' dan 'Female'.

4. Kolom Item Purchased: Berisi jenis barang yang dibeli pelanggan, dengan nilai unik seperti 'Blouse', 'Sweater', 'Jeans', hingga 'Gloves'.

5. Kolom Category: Menunjukkan kategori barang yang dibeli, dengan nilai unik berupa 'Clothing', 'Footwear', 'Outerwear', dan 'Accessories'.

6. Kolom Purchase Amount (USD): Berisi jumlah pembelian dalam dolar AS, dengan nilai unik mulai dari $20 hingga $100.

7. Kolom Location: Menunjukkan lokasi pelanggan berdasarkan negara bagian di AS, dengan nilai unik seperti 'Kentucky', 'Maine', hingga 'Michigan'.

8. Kolom Size: Menunjukkan ukuran produk yang dibeli, dengan nilai unik berupa 'L', 'S', 'M', dan 'XL'.

9. Kolom Color: Berisi warna produk yang dibeli, dengan nilai unik seperti 'Gray', 'Maroon', 'Blue', hingga 'Magenta'.

10. Kolom Season: Menunjukkan musim saat pembelian terjadi, dengan nilai unik berupa 'Winter', 'Spring', 'Summer', dan 'Fall'.

11. Kolom Review Rating: Berisi skor ulasan produk dengan skala dari 2.5 hingga 5.0, menunjukkan tingkat kepuasan pelanggan.

12. Kolom Subscription Status: Menunjukkan status langganan pelanggan, dengan nilai unik berupa 'Yes' dan 'No'.

13. Kolom Payment Method: Menunjukkan metode pembayaran yang digunakan pelanggan, dengan nilai unik seperti 'Credit Card', 'Bank Transfer', 'Cash', hingga 'Debit Card'.

14. Kolom Shipping Type: Berisi jenis pengiriman yang dipilih pelanggan, dengan nilai unik berupa 'Express', 'Free Shipping', 'Next Day Air', hingga 'Store Pickup'.

15. Kolom Discount Applied: Menunjukkan apakah diskon diterapkan, dengan nilai unik berupa 'Yes' dan 'No'.

16. Kolom Promo Code Used: Menunjukkan apakah kode promo digunakan, dengan nilai unik berupa 'Yes' dan 'No'.

17. Kolom Previous Purchases: Berisi jumlah pembelian sebelumnya oleh pelanggan, dengan nilai unik berkisar dari 1 hingga 50.

18. Kolom Preferred Payment Method: Menunjukkan metode pembayaran yang paling sering digunakan pelanggan, dengan nilai unik seperti 'Venmo', 'Cash', 'Credit Card', hingga 'Debit Card'.

19. Kolom Frequency of Purchases: Menunjukkan frekuensi pembelian pelanggan, dengan nilai unik seperti 'Fortnightly', 'Weekly', 'Monthly', hingga 'Every 3 Months'.

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
# Memisahkan kolom numerik dan kolom kategorikal

numerikal = X.select_dtypes(include=['number']).columns
kategorikal = X.select_dtypes(include=['object']).columns

# Kolom numerik diproses dengan standarisasi.
# Kolom kategorikal diproses dengan one-hot encoding.
```

```python
# Membuat Pipeline untuk Data Numerik
numerical_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())])   # Skala data numerik

# Pipeline numerik hanya melakukan standarisasi.
# Pipeline kategorikal hanya melakukan encoding.
```

```python
# Membuat Pipeline untuk Data Kategorikal
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))  # Encoding One-Hot untuk data kategorikal
])
```

```python
#  Menggabungkan Transformasi Menggunakan ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerikal),  # Proses data numerik
        ('cat', categorical_transformer, kategorikal)  # Proses data kategorikal
    ])
# Pipeline numerik dan kategorikal diterapkan sesuai jenis kolom.
# Output adalah dataset yang sudah diproses, siap digunakan untuk model machine learning.
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
```

```python
pipelines = {
    'Logistic Regression': Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', LogisticRegression())
    ]),
    'Decision Tree': Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', DecisionTreeClassifier())
    ]),
    'Random Forest': Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', RandomForestClassifier())
    ]),
    'Gradient Boosting': Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', GradientBoostingClassifier())
    ])
}
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Dictionary untuk menyimpan hasil evaluasi
evaluation_results = {}

# Fungsi untuk menghitung metrik evaluasi
def evaluate_model(y_true, y_pred):
    return {
        'Accuracy': accuracy_score(y_true, y_pred),
        'Precision': precision_score(y_true, y_pred, average='weighted'),
        'Recall': recall_score(y_true, y_pred, average='weighted'),
        'F1 Score': f1_score(y_true, y_pred, average='weighted')
    }
    # Latih dan evaluasi setiap model
for name, pipeline in pipelines.items():
    pipeline.fit(X_train, y_train)  # Latih model
    y_train_pred = pipeline.predict(X_train)
    y_test_pred = pipeline.predict(X_test)
    
    # Simpan hasil evaluasi pada train dan test
    evaluation_results[name] = {
        'Train': evaluate_model(y_train, y_train_pred),
        'Test': evaluate_model(y_test, y_test_pred)
    }
    
    # Cetak hasil evaluasi
    train_metrics = evaluation_results[name]['Train']
    test_metrics = evaluation_results[name]['Test']
    print(f"{name}:")
    print(f"  Train: {train_metrics}")
    print(f"  Test: {test_metrics}")
    print()
```

```python
# Konversi hasil evaluasi menjadi DataFrame
results_table = []

for model_name, metrics in evaluation_results.items():
    results_table.append({
        'Model': model_name,
        'Train Accuracy': metrics['Train']['Accuracy'],
        'Train Precision': metrics['Train']['Precision'],
        'Train Recall': metrics['Train']['Recall'],
        'Train F1 Score': metrics['Train']['F1 Score'],
        'Test Accuracy': metrics['Test']['Accuracy'],
        'Test Precision': metrics['Test']['Precision'],
        'Test Recall': metrics['Test']['Recall'],
        'Test F1 Score': metrics['Test']['F1 Score']
    })

# Buat DataFrame untuk tabel
df_results = pd.DataFrame(results_table)


# Urutkan berdasarkan 'Test F1 Score' dari yang terbesar ke terkecil
df_results_imbalace = df_results.sort_values(by='Test F1 Score', ascending=False)

# Tampilkan tabel yang sudah diurutkan
print(df_results_imbalace)
```

* Gradient Boosting: Meski tidak sepenuhnya terhindar dari overfitting, Gradient Boosting menunjukkan kinerja yang lebih stabil antara data latih dan uji, meskipun ada penurunan performa di data uji.

# Menggunakan SMOTE

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
# Merapkan SMOTE untuk menangani imbalance
smote = SMOTE(sampling_strategy='auto', random_state=42)

# Terapkan SMOTE hanya pada data training (pastikan tidak dilakukan pada data test)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
# Langkah 4: Bangun pipeline untuk preprocessing dan berbagai model

# List model yang ingin digunakan
models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42)
}

# Dictionary untuk menyimpan hasil evaluasi
evaluation_results = {}

# Loop melalui model dan latih dengan data yang sudah di-resample
for model_name, model in models.items():
    pipeline = Pipeline([
        ('preprocessor', ColumnTransformer(
            transformers=[
                ('num', StandardScaler(), X.select_dtypes(include=['number']).columns),
                ('cat', OneHotEncoder(handle_unknown='ignore'), X.select_dtypes(include=['object']).columns)
            ]
        )),
        ('classifier', model)
    ])

    # Langkah 5: Latih model dengan data yang sudah di-resample
    pipeline.fit(X_train_resampled, y_train_resampled)

    # Langkah 6: Prediksi dan evaluasi model
    y_train_pred = pipeline.predict(X_train)
    y_test_pred = pipeline.predict(X_test)
    
    # Evaluasi metrik pada data train
    train_accuracy = accuracy_score(y_train, y_train_pred)
    train_precision = precision_score(y_train, y_train_pred, average='weighted')
    train_recall = recall_score(y_train, y_train_pred, average='weighted')
    train_f1 = f1_score(y_train, y_train_pred, average='weighted')
    
    # Evaluasi metrik pada data test
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_precision = precision_score(y_test, y_test_pred, average='weighted')
    test_recall = recall_score(y_test, y_test_pred, average='weighted')
    test_f1 = f1_score(y_test, y_test_pred, average='weighted')
    
    # Simpan hasil ke dictionary
    evaluation_results[model_name] = {
        'Train': {
            'Accuracy': train_accuracy,
            'Precision': train_precision,
            'Recall': train_recall,
            'F1 Score': train_f1
        },
        'Test': {
            'Accuracy': test_accuracy,
            'Precision': test_precision,
            'Recall': test_recall,
            'F1 Score': test_f1
        }
    }
    
    # Cetak hasil evaluasi
    print(f"{model_name}:")
    print(f"  Train: Accuracy={train_accuracy:.4f}, Precision={train_precision:.4f}, Recall={train_recall:.4f}, F1 Score={train_f1:.4f}")
    print(f"  Test: Accuracy={test_accuracy:.4f}, Precision={test_precision:.4f}, Recall={test_recall:.4f}, F1 Score={test_f1:.4f}")
    print()
```

```python
# Konversi hasil evaluasi menjadi DataFrame
results_table = []

for model_name, metrics in evaluation_results.items():
    results_table.append({
        'Model SMOTE': model_name,  # Ganti 'Model' menjadi 'Model SMOTE'
        'Train Accuracy': metrics['Train']['Accuracy'],
        'Train Precision': metrics['Train']['Precision'],
        'Train Recall': metrics['Train']['Recall'],
        'Train F1 Score': metrics['Train']['F1 Score'],
        'Test Accuracy': metrics['Test']['Accuracy'],
        'Test Precision': metrics['Test']['Precision'],
        'Test Recall': metrics['Test']['Recall'],
        'Test F1 Score': metrics['Test']['F1 Score']
    })

# Buat DataFrame untuk tabel
df_results = pd.DataFrame(results_table)

# Urutkan berdasarkan 'Test F1 Score' dari yang terbesar ke terkecil
df_results_sorted = df_results.sort_values(by='Train F1 Score', ascending=False)

# Tampilkan tabel yang sudah diurutkan
print(df_results_sorted)
```

* Random Forest dan Decision Tree masih menunjukkan overfitting yang jelas, dengan akurasi dan F1 score 100% di data latih, tetapi ada penurunan signifikan pada data uji. Meskipun recall masih cukup baik, model ini bisa lebih baik jika diterapkan regularisasi atau tuning hyperparameter untuk mengurangi overfitting.
* Gradient Boosting dan Logistic Regression menunjukkan keseimbangan yang lebih baik antara precision dan recall, meskipun masih ada ruang untuk perbaikan. Precision dan recall yang hampir setara menunjukkan bahwa model mampu mengidentifikasi kedua kelas (berlangganan dan tidak berlangganan) dengan cukup baik.

# Hyperparameter Tuning

Random Forest dan Decision Tree mungkin perlu disesuaikan dengan menggunakan teknik pruning) untuk mengurangi overfitting.

```python
import optuna
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline

# Fungsi untuk mendefinisikan tujuan dari pencarian parameter
def objective(trial, model_name):
    # Tentukan parameter yang ingin di-tune untuk setiap model
    if model_name == 'Logistic Regression':
        C = trial.suggest_loguniform('C', 1e-5, 100)  # Regularization strength
        solver = trial.suggest_categorical('solver', ['liblinear', 'lbfgs'])  # Solver untuk Logistic Regression
        model = LogisticRegression(C=C, solver=solver)
    
    elif model_name == 'Random Forest':
        n_estimators = trial.suggest_int('n_estimators', 50, 200)  # Jumlah pohon
        max_depth = trial.suggest_int('max_depth', 5, 20)  # Kedalaman pohon
        model = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth)
    
    elif model_name == 'Decision Tree':
        max_depth = trial.suggest_int('max_depth', 5, 20)  # Kedalaman pohon
        min_samples_split = trial.suggest_int('min_samples_split', 2, 20)  # Minimal sampel untuk split
        model = DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split)

    # Buat pipeline dengan preprocessor dan model
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),  # Preprocessing data
        ('classifier', model)
    ])
    
    # Latih model
    pipeline.fit(X_train, y_train)
    
    # Prediksi dan hitung akurasi pada data test
    y_pred = pipeline.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    
    return accuracy

# Definisikan model yang akan di-tune
models = ['Logistic Regression', 'Random Forest', 'Decision Tree']

# Cari hyperparameter terbaik untuk setiap model menggunakan Optuna
for model_name in models:
    study = optuna.create_study(direction='maximize')  # Mencari akurasi tertinggi
    study.optimize(lambda trial: objective(trial, model_name), n_trials=50)  # Lakukan 50 trial
    
    print(f"Best hyperparameters for {model_name}:")
    print(study.best_params)
    print(f"Best accuracy: {study.best_value:.4f}")
    print()
```

```python
# Daftar model yang akan di-tune
models = ['Random Forest', 'Decision Tree']

# Membuat hasil pencarian untuk setiap model
best_params = {}

for model_name in models:
    # Tentukan study untuk setiap model
    study = optuna.create_study(direction='maximize')  # optimize untuk meningkatkan akurasi
    study.optimize(lambda trial: objective(trial, model_name), n_trials=10)  # 50 percobaan

    # Simpan hasil terbaik untuk model ini
    best_params[model_name] = study.best_params

    print(f"Best parameters for {model_name}: {study.best_params}")
    print(f"Best accuracy: {study.best_value}")
```
